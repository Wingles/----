```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = 'SimHei'

# 读取数据
data = pd.read_excel('数据报表.xlsx', sheet_name=0)

# 准备特征和目标变量
X = data[['年龄']]
y = data['消费金额']

# 创建线性回归模型
model = LinearRegression()

# 拟合模型
model.fit(X, y)

# 进行预测
age_array = [10, 20, 30, 40, 50, 60, 70]  # 年龄数组
predicted_amounts = model.predict(pd.DataFrame(age_array, columns=['年龄']))

# 打印预测结果
for age, amount in zip(age_array, predicted_amounts):
    print(f"年龄 {age} 的预测消费金额为: {amount}")

# 可视化预测结果
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='实际数据')
plt.plot(age_array, predicted_amounts, color='red', linewidth=2, label='预测线')
plt.xlabel('年龄')
plt.ylabel('消费金额')
plt.title('年龄与消费金额的线性回归预测')
plt.legend()
plt.show()
```

```python
# 准备特征和目标变量
X = data[['消费体验']]
y = data['消费金额']

# 将消费体验文本值转换为数值表示
X.replace({'好': 5, '较好': 4, '一般': 3, '较差': 2, '差': 1}, inplace=True)

# 创建线性回归模型
model = LinearRegression()

# 拟合模型
model.fit(X, y)

# 进行预测
predicted_amounts = model.predict(X)

# 可视化实际数据和预测结果
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='实际数据')
plt.plot(X, predicted_amounts, color='red', linewidth=2, label='预测线')
plt.xlabel('消费体验')
plt.ylabel('消费金额')
plt.title('消费金额与消费体验的线性回归预测')
plt.legend()
plt.show()
```
