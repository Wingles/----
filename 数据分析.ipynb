```python
# 从Excel中读取数据
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = 'SimHei'
data = pd.read_excel('数据报表.xlsx',sheet_name=0)
```

```python
# 对年龄和消费金额进行统计学分析
amount = data['消费金额']
age = data['年龄']

# 计算统计指标
amount_mean = np.mean(amount)
amount_median = np.median(amount)
age_mean = np.mean(age)
age_median = np.median(age)

# 创建散点图
plt.scatter(age,amount,alpha=0.5)
plt.xlabel('年龄')
plt.ylabel('消费金额')
plt.title('年龄与消费金额关系')

# 显示统计指标
plt.text(0.1, 0.9, f'消费金额均值: {amount_mean:.2f}', transform=plt.gca().transAxes)
plt.text(0.1, 0.85, f'消费金额中位数: {amount_median:.2f}', transform=plt.gca().transAxes)
plt.text(0.1, 0.8, f'年龄均值: {age_mean:.2f}', transform=plt.gca().transAxes)
plt.text(0.1, 0.75, f'年龄中位数: {age_median:.2f}', transform=plt.gca().transAxes)

# 显示图形
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 假设已有数据集，其中包含 '年龄' 和 '主要类别' 列
category = data['主要类别']
age = data['年龄']
df = pd.DataFrame({'年龄': age, '主要类别': category})

# 按照主要类别进行分组并计算年龄的统计指标
grouped = df.groupby('主要类别')
category_stats = grouped.describe()

# 创建箱线图
plt.figure(figsize=(8, 6))
df.boxplot(column='年龄', by='主要类别')
plt.xlabel('主要类别')
plt.ylabel('年龄')
plt.title('年龄与主要类别的关系')

# 显示统计指标
stats_text = ''
for category, stats in category_stats['年龄'].items():
    stats_text += f'{category}:\n'
    for i, (stat, value) in enumerate(stats.items()):
        if i >= 5:  # 限制显示5个指标
            break
        stats_text += f'  {stat}: {value}\n'
    stats_text += '\n'
plt.text(0.05, 3.5, stats_text, transform=plt.gca().transAxes, fontsize=10, verticalalignment='top')

# 显示图形
plt.show()
```

```python
count_by_method = data['主要方式'].value_counts()

# 统计每种主要方式在不同年龄段的分布
age_by_method = data.groupby('主要方式')['年龄'].mean()

# 创建柱状图可视化不同主要方式的人数
count_by_method.plot(kind='bar', rot=0)
plt.xlabel('主要方式')
plt.ylabel('人数')
plt.title('不同主要方式的人数分布')
plt.show()


```
